```python
# Write code to find mean, mode of the number of runs scored by top 10 players. IPL dataset is available in classroom
import pandas as pd

df = pd.read_csv('ipl_matches_smalldata.csv')

print(df.head())

top_10_players = df[['batter', 'total_runs']].sort_values(by='total_runs', ascending=False).head(10)

mean_runs = top_10_players['total_runs'].mean()
mode_runs = top_10_players['total_runs'].mode()

print(f"Mean of runs scored by top 10 players: {mean_runs}")
print(f"Mode of runs scored by top 10 players: {mode_runs}")
```

```python
# Write code to filter players who have scored more than 500 runs
import pandas as pd

df = pd.read_csv('ipl_matches_smalldata.csv')

print(df.head())

players_above_500_runs = df[df['total_runs'] > 500]

print(players_above_500_runs)
```

```python
Writ # code to create a bar chart to visualize the number of runs by top 10 players
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ipl_matches_smalldata.csv')

print(df.head())

top_10_players = df[['batter', 'total_runs']].sort_values(by='total_runs', ascending=False).head(10)


plt.figure(figsize=(10, 6))
plt.bar(top_10_players['batter'], top_10_players['total_runs'], color='skyblue')

plt.xlabel('batter')
plt.ylabel('total_runs')
plt.title('Top 10 Players by Runs Scored')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()  
plt.show()
```

```python

# Write a Python program using OpenCV functions to 
# 1. read an image from a file and display it using OpenCV.
# 2. Find the shape (dimensions) of the image. 
# 3. Resize the image to dimensions of 400x256 pixels.
# 4.Display the resized image.
# 5.Flip the image horizontally and vertically
import cv2

image = cv2.imread('image.PNG')  
if image is None:
    print("Error: Image not found.")
else:

    cv2.imshow('Original Image', image)
    cv2.waitKey(0)  

    height, width, channels = image.shape
    print(f"Original Image Dimensions: {width}x{height} (Width x Height)")

    resized_image = cv2.resize(image, (400, 256))

    cv2.imshow('Resized Image', resized_image)
    cv2.waitKey(0)

    flipped_horizontally = cv2.flip(resized_image, 1) 
    flipped_vertically = cv2.flip(resized_image, 0)  


    cv2.imshow('Flipped Horizontally', flipped_horizontally)
    cv2.imshow('Flipped Vertically', flipped_vertically)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# Write a python program to load dataset related to tips given in restaurant and display a pie chart to compare percentage of people in smoking area vs non smoking area
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('tips.csv')  

print(df.head())

smoking_area_count = df['Smoking'].value_counts()

labels = smoking_area_count.index  
sizes = smoking_area_count.values  
colors = ['lightcoral', 'lightgreen']  

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, textprops={'fontsize': 12})
plt.title('Percentage of People in Smoking Area vs Non-Smoking Area', fontsize=14)
plt.axis('equal') 

plt.show()
```

```python
# Write python program using nltk library to load the text on topic of your choice and generate a word cloud from it.
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

nltk.download('punkt')  
nltk.download('stopwords') 

text = """
Artificial Intelligence (AI) is intelligence demonstrated by machines, as opposed to the natural intelligence displayed by humans and animals.
Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals.
As machines become increasingly capable, tasks considered to require "intelligence" are often removed from the definition of AI, a phenomenon known as the AI effect.
AI research has attempted to emulate human cognition. The field was founded on the claim that "any aspect of learning or any other feature of intelligence can in principle be so precisely described that a machine can be made to simulate it."
"""

words = word_tokenize(text)

words = [word.lower() for word in words if word.isalpha()]

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

filtered_text = ' '.join(filtered_words)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Topic: Artificial Intelligence')
plt.show()
```
